6. Look at the dependency and PTB representations for an active sentence and its corresponding passive sentence. e.g. `The dog ate the pie.` or `The pie was eaten by the dog.`. Do either of these parses give hints about the semantic link between active/passive pairs? [5 pts]

    __THE DOG ATE THE PIE.__

    __PTB__

        (ROOT
          (S
            (NP (DT The) (NN dog))
            (VP (VBD ate)
              (NP (DT the) (NN pie)))
            (. .)))

    __Dependency Parser__

        det(dog-2, The-1)
        nsubj(ate-3, dog-2)
        root(ROOT-0, ate-3)
        det(pie-5, the-4)
        dobj(ate-3, pie-5)

    __THE PIE WAS EATEN BY THE DOG.__

    __PTB__

        (ROOT
          (S
            (NP (DT The) (NN pie))
            (VP (VBD was)
              (VP (VBN eaten)
                (PP (IN by)
                  (NP (DT the) (NN dog)))))
            (. .)))

    __Dependency Parser__

        det(pie-2, The-1)
        nsubjpass(eaten-4, pie-2)
        auxpass(eaten-4, was-3)
        root(ROOT-0, eaten-4)
        case(dog-7, by-5)
        det(dog-7, the-6)
        nmod(eaten-4, dog-7)

    __COMMENTS__

    When we compare the dependency parser outputs of the active and the passive sentences, we see that there is the following dependency between `eat/ate/eaten` and `pie` in the active sentence; `dobj(ate-3, pie-5)` and a corresponding dependency in the passive sentence; `nsubjpass(eaten-4, pie-2)`. This makes sense as the objects of active sentences should be appearing as subjects of corresponding passive sentences because of the way that passive sentences are constructed.

    On the other hand, we can see in the PTB parser output that the phrase and word tags remain the same and only the order of the output changes. This, by itself can't guarantee that these sentences are semantically close, but when we consider the PTB output together with the dependencies, we can clearly see that these sentences are close in meaning.

8. Give a rule (informally, in English) that will allow you to use a dependency parse to detect number agreement errors (i.e., The cat eat fish.) [10 pts]

    __Answer:__ When we use a dependency parser with this example (The cat eat fish.), we will obtain the following relation: `nsubj(eat, cat)`. We can detect this type of agreement errors as follows:
    - If the verb is in simple present tense in first or second person (do, go, eat) and the subject is a singular third person\object pronoun (he/she/it) or a singular noun, then the parser should return a number agreement error. 
    - Otherwise, if the verb is in present simple in first or second person (do, go, eat) and the subject is a singular first or second person pronoun or any plural pronoun/noun, then the parser should run without number agreement errors.

10. Consider a model trained on data annotated with dependencies or PTB structure (like the online parsers you’ve been using), and an NLP task (e.g. translation, NLI). Would the model do better or worse on that task than a model trained on data with no syntactic annotation? Why? (no right answer here, just give at least one argument for your reasoning) [5 pts]

    __Answer__: I would expect a model trained on data annotated with dependencies and/or PTB structure to perform same or worse as an encoder for translation/NLI, because I do not think that using dependency/PTB tags will make substantial contribution to the generalization ability of the model. Imagine we annotate the data with dependency and train our model, I believe there is a good chance that the model will favor the dependencies it has seen, just like the n-gram classifier which is better at handling known n-grams than it is at handling unknown/unseen ones. When we train a model without dependency/PTB annotation, it can also capture dependencies between words in a sentence without requiring any supervision. When we annotate the data on the other hand, we are forcing the model to learn those dependencies, that might not be necessarily crucial for the task. One might also argue that, if they are indeed not crucial, a neural network will learn this as well, without requiring any supervision, but this itself does not guarantee the performance will increase on the test set. Another downside of annotating the data with dependency/PTB info is that annotating sentences that are ambiguous is hard, and the parser/annotator can easily get the meaning wrong in some cases. As in all cases, with infinite data, dependency/PTB annotation could outperform other methods. However, there are so many distinct rules in human language that I believe we would run out of time before we could account for enough of them by extracting features from parsers.
